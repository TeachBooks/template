$R = factor *x* F$

The numbers in the tables are related to the concentrated loads working on the structure. The bending moments and support reactions in the designated points are found by multiplying the factors in the table with F x L or F.

For example:
- M(1) is the maximum bending in beam (1).
- M3 is the hogging moment located at support 3.
- R2 is the support reaction in support 2.

- F = 30 kN
- L = 8 m

The calculations are as follows:
- M(1) = 0.175 x 30 x 8 kNm
- M3 = -0.15 x 30 x 8 kNm
- R2 = 1.15 x 30 kN

$M = factor * q * L^2$  
$R = factor * q * L$

The numbers in the tables are related to the concentrated loads
working on the structure. The bending moments and support reactions in
the designated points are found by multiplying the factors in the table with
$q * L^2$ or $q * L$.

For example: M(1) is the maximum bending in beam
(1). M3 is the hogging moment located at
support 3. R2 is the support reaction in support 2.  
  
For situation m:  
$F = 5 kN/m$
$L = 6 m$  
$M(1) = 0,0735 * 5 * 62 kNm$  
$M3 = -0,0333 * 5 * 62 kNm$  
$R2 = 1,2001 * 5 * 6 kN$  
  
*) location of bending moment M(1) in relation to support 1 (distance = factor x L)  
**) location of bending moment M(2) in relation to support 2 (distance = factor
